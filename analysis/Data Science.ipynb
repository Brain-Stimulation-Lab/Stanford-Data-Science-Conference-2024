```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
sns.set_style("whitegrid") # Set style for chart

bars = [ 'No code publically available', 'Code available',]
counts = [39, 11]
bar_labels = ['No code', 'Code' ]
bar_colors = [ '#cfe7f0', '#eeed95']
ax.grid(False)

ax.bar(bars, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Count')
ax.legend(title='Papers surveyed')

plt.rcParams.update({'font.size': 16})

# Save the figure
plt.savefig("barchart.png")

plt.show()
```
