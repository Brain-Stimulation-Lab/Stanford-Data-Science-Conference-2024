```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
sns.set_style("whitegrid") # Set style for chart

fruits = ['Code available', 'No code publically available']
counts = [11, 39]
bar_labels = ['Code', 'No code', ]
bar_colors = ['#eeed95', '#cfe7f0']
ax.grid(False)

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Count')
ax.legend(title='Papers surveyed')

plt.rcParams.update({'font.size': 16})

# Save the figure
plt.savefig("barchart.png")

plt.show()
```
